```python
from scipy.stats import spearmanr, pearsonr
import pandas as pd
```

## NNScore

```python
all_pdb=pd.read_csv("../split_dataset/cluster/pdbbind_2020_cluster_result.csv")
cluster_summary=pd.read_csv("../split_dataset/cluster/PCV_cluster_summary.csv")
need_clu = cluster_summary[cluster_summary['pdb_num']>20]['cluster'].to_list()
need_clu.remove("Pkinase")
need_clu.append('pkinase')
weights=pd.read_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/cluster_weights.csv",)
```

```python
nnscore=pd.read_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/NN_RCV_10_all_pdb_score.csv")
nnscore_rcv=[]
for i in range(1,11):
    for j in need_clu:
        x=nnscore[nnscore['PCV_cluster']==j]['pred_model_'+str(i)]
        y=nnscore[nnscore['PCV_cluster']==j]['affinity']
        sp = spearmanr(x, y)
        nnscore_rcv.append([i, j, sp[0], sp[1],'NNScore','Random-CV'])
        # print(sp)
nnscore_rcv=pd.DataFrame(nnscore_rcv, columns=['i','cluster','spearmanr','p-value','model','CV'])
# nnscore_rcv.to_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/nnscore_rcv.csv", index=False)
nnscore_rcv=pd.merge(nnscore_rcv, weights, on=['cluster'])
nnscore_rcv.to_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/nnscore_rcv.csv", index=False)
nnscore=pd.read_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/NN_SCV_10_all_pdb_score.csv")
nnscore_scv=[]
for i in range(1,11):
    for j in need_clu:
        x=nnscore[nnscore['PCV_cluster']==j]['pred_model_'+str(i)]
        y=nnscore[nnscore['PCV_cluster']==j]['affinity']
        sp = spearmanr(x, y)
        nnscore_scv.append([i, j, sp[0], sp[1],'NNScore','Seq-CV'])
        # print(sp)
nnscore_scv=pd.DataFrame(nnscore_scv, columns=['i','cluster','spearmanr','p-value','model','CV'])
# nnscore_scv.to_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/nnscore_scv.csv", index=False)
nnscore_scv=pd.merge(nnscore_scv, weights, on=['cluster'])
nnscore_scv.to_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/nnscore_scv.csv", index=False)
nnscore=pd.read_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/NN_PCV_10_all_pdb_score.csv")
nnscore_pcv=[]
for i in range(1,11):
    for j in need_clu:
        x=nnscore[nnscore['PCV_cluster']==j]['pred_model_'+str(i)]
        y=nnscore[nnscore['PCV_cluster']==j]['affinity']
        sp = spearmanr(x, y)
        nnscore_pcv.append([i, j, sp[0], sp[1],'NNScore','Pfam-CV'])
        # print(sp)
nnscore_pcv=pd.DataFrame(nnscore_pcv, columns=['i','cluster','spearmanr','p-value','model','CV'])
# nnscore_pcv.to_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/nnscore_pcv.csv", index=False)
nnscore_pcv=pd.merge(nnscore_pcv, weights, on=['cluster'])
nnscore_pcv.to_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/nnscore_pcv.csv", index=False)
nnscore_out = []
for i in range(1,11):
    scv_score=(nnscore_scv[nnscore_scv['i']==i]['spearmanr']*nnscore_scv[nnscore_scv['i']==i]['weight']).sum()
    rcv_score=(nnscore_rcv[nnscore_rcv['i']==i]['spearmanr']*nnscore_rcv[nnscore_rcv['i']==i]['weight']).sum()
    pcv_score=(nnscore_pcv[nnscore_pcv['i']==i]['spearmanr']*nnscore_pcv[nnscore_pcv['i']==i]['weight']).sum()
    nnscore_out.append([i, scv_score,'NNScore','Seq-CV'])
    nnscore_out.append([i, rcv_score,'NNScore','Random-CV'])
    nnscore_out.append([i, pcv_score,'NNScore','Pfam-CV'])

nnscore_out=pd.DataFrame(nnscore_out, columns=['i','weighted_spearman','model','CV'])
nnscore_out.to_csv("/pubhome/hzhu02/GPSF/generalization_benchmark/models/general_3_fold_summary/data_mw_bias/spearman/nnscore_out.csv", index=False)
```
